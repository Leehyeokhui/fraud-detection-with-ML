# 이상치 확인

```python
def count_outlier(data, columns):
    df_o = pd.DataFrame(index=['U-out','L-out'], columns=columns)
    for column in columns:
        q25, q75 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)
        iqr = q75 - q25
        lower, upper = q25 - iqr*1.5, q75 + iqr*1.5
        U_outlier = len(data[data[column] > upper])
        L_outlier = len(data[data[column] < lower])
        df_o[column][0] = U_outlier
        df_o[column][1] = L_outlier
    return df_o
```

```python
df_o = count_outlier(df_x, df_x.columns)
df_o
```
